## Part 1

```python
# Parse data into a 2d grid
grid = [list(row) for row in data.split("\n")]
```

```python
distinct_positions = set()
```

```python
# Find the start point of the grid and it's direction. Then replace it with '.'
current_point = None
for y, row in enumerate(grid):
    for x, cell in enumerate(row):
        if cell in "^v<>":
            current_point = (x, y)
            if cell == "^":
                direction = (0, -1)
            elif cell == "v":
                direction = (0, 1)
            elif cell == "<":
                direction = (-1, 0)
            elif cell == ">":
                direction = (1, 0)
            grid[y][x] = "."
            break
    
    if current_point:
        break

# Then, add it's start point to distinct positions
distinct_positions.add(current_point)
```

```python
def rotate_direction(direction):
    if direction == (0, -1):
        return (1, 0)
    elif direction == (1, 0):
        return (0, 1)
    elif direction == (0, 1):
        return (-1, 0)
    elif direction == (-1, 0):
        return (0, -1)
```

```python
while True:
    # Fill in next cell on the grid with 'X' and print out
    # grid[current_point[1]][current_point[0]] = 'X'
    # print('------------------')
    # for row in grid:
    #     print("".join(row))
    # print()

    finished = False
    x, y = current_point
    # First check the next cell in the direction we are facing
    # if it's 'X', then we need to rotate and move forward
    # if it's out of bounds, we finish
    while True:
        next_cell = (x + direction[0], y + direction[1])
        if next_cell[0] < 0 or next_cell[0] >= len(grid[0]) or next_cell[1] < 0 or next_cell[1] >= len(grid):
            finished = True
            break
        elif grid[next_cell[1]][next_cell[0]] == '#':
            direction = rotate_direction(direction)
        else:
            break
    
    if finished:
        break

    distinct_positions.add(next_cell)

    current_point = next_cell
```

```python
# I'm going to do this the super hacky way: just test every position, keep track of the iterations of the simulator, and if it ever becomes more than than the size of the grid,
# we know it's in a loop

# Get size of grid
grid_width = len(grid[0])
grid_height = len(grid)
size = grid_width * grid_height
print(grid_height)
print(grid_width)
print(size)
```

```python
def run_simulator(simulator_grid, current_point, direction):
    num_movements = 0
    while num_movements < 1000000:
        x, y = current_point
        # First check the next cell in the direction we are facing
        # if it's 'X', then we need to rotate and move forward
        # if it's out of bounds, we finish
        while True:
            next_cell = (x + direction[0], y + direction[1])
            if next_cell[0] < 0 or next_cell[0] >= len(simulator_grid[0]) or next_cell[1] < 0 or next_cell[1] >= len(simulator_grid):
                return True
            elif simulator_grid[next_cell[1]][next_cell[0]] == '#':
                direction = rotate_direction(direction)
            else:
                break

        current_point = next_cell

        num_movements += 1

    return False
```
