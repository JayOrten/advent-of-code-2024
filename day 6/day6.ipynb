```python
# Parse data into a 2d grid
grid = [list(row) for row in data.split("\n")]
```

## Part 1

```python
distinct_positions = set()
```

```python
# Find the start point of the grid and it's direction. Then replace it with '.'
current_point = None
for row_index, row in enumerate(grid):
    for column_index, cell in enumerate(row):
        if cell in "^v<>":
            current_point = (row_index, column_index)
            if cell == "^":
                direction = (-1, 0)
            elif cell == "v":
                direction = (1, 0)
            elif cell == "<":
                direction = (0, -1)
            elif cell == ">":
                direction = (0, 1)
            grid[row_index][column_index] = "."
            break
    
    if current_point:
        break

# Then, add it's start point to distinct positions
distinct_positions.add(current_point)
```

```python
def rotate_direction(direction):
    if direction == (-1, 0): # Up
        return (0, 1)
    elif direction == (0, 1): # Right
        return (1, 0)
    elif direction == (1, 0): # Down
        return (0, -1)
    elif direction == (0, -1): # Left
        return (-1, 0)
```

```python
while True:
    finished = False
    row_index, column_index = current_point
    # First check the next cell in the direction we are facing
    # if it's 'X', then we need to rotate and move forward
    # if it's out of bounds, we finish
    while True:
        next_cell = (row_index + direction[0], column_index + direction[1])
        if next_cell[0] < 0 or next_cell[0] >= len(grid) or next_cell[1] < 0 or next_cell[1] >= len(grid[0]):
            finished = True
            break
        elif grid[next_cell[0]][next_cell[1]] == '#':
            direction = rotate_direction(direction)
        else:
            break
    
    if finished:
        break

    distinct_positions.add(next_cell)

    current_point = next_cell
```

```python
# Find the start point of the grid and it's direction. Then replace it with '.'
current_point = None
for row_index, row in enumerate(grid):
    for column_index, cell in enumerate(row):
        if cell in "^v<>":
            current_point = (row_index, column_index)
            if cell == "^":
                direction = (-1, 0)
            elif cell == "v":
                direction = (1, 0)
            elif cell == "<":
                direction = (0, -1)
            elif cell == ">":
                direction = (0, 1)
            grid[row_index][column_index] = "."
            break
    
    if current_point:
        break
```

```python
def run_simulator(simulator_grid, current_point, direction):
    num_movements = 0
    while num_movements < 100000:
        row_index, column_index = current_point
        # First check the next cell in the direction we are facing
        # if it's 'X', then we need to rotate and move forward
        # if it's out of bounds, we finish
        while True:
            next_cell = (row_index + direction[0], column_index + direction[1])
            if next_cell[0] < 0 or next_cell[0] >= len(simulator_grid) or next_cell[1] < 0 or next_cell[1] >= len(simulator_grid[0]):
                return True
            elif simulator_grid[next_cell[0]][next_cell[1]] == '#':
                direction = rotate_direction(direction)
            else:
                break

        current_point = next_cell
        simulator_grid[current_point[0]][current_point[1]] = 'X'

        num_movements += 1

    return False
```

```python
num_loops = 0
```

```python
for row_index in range(grid_height):
    for column_index in range(grid_width):
        print("checking", row_index, column_index)
        # If it's the start position, skip
        if (row_index, column_index) == current_point:
            continue

        simulator_grid = copy.deepcopy(grid)
        simulator_grid[row_index][column_index] = '#'
        
        if not run_simulator(simulator_grid, current_point, direction):
            num_loops += 1
```
