Part 1

```python
from aocd import data
```

```python
import re
```

```python
results = [x.group() for x in re.finditer( r'mul\((\d{1,3}),(\d{1,3})\)', data)]
```

```python
multiplication_sum = 0
```

```python
for mul in results:
    nums = re.findall( r'mul\((.*?),(.*?)\)', mul)[0]
    a = int(nums[0])
    b = int(nums[1])
    product = a * b
    multiplication_sum += product
```

```python
multiplication_sum
```

Part 2

```python
# Grab the indices of the results, to check ordering for activation and deactivation
result_indices = [x.start() for x in re.finditer( r'mul\((\d{1,3}),(\d{1,3})\)', data)]
```

```python
# Pair together result indices with the results
result_pairs = list(zip(result_indices, results))
```

```python
# Get indices of do and don't
do_indices = [0] + [x.start() for x in re.finditer( r'do\(\)', data)]
do_indices = [(x, 'do') for x in do_indices]
```

```python
dont_indices = [x.start() for x in re.finditer( r'don\'t\(\)', data)]
dont_indices = [(x, 'dont') for x in dont_indices]
```

```python
# Combine the lists and sort them
combined_indices = do_indices + dont_indices
combined_indices.sort()
```

```python
# Iterate through and remove duplicates
for i in range(len(combined_indices)-1, -1, -1):
    if combined_indices[i][1] == combined_indices[i-1][1]:
        del combined_indices[i]
```

```python
# It deletes the first one because it wraps around, but that's okay. We can just add it back in
combined_indices.insert(0, (0, 'do'))
# Also, add a don't at one million to the end
combined_indices.append((1000000, 'dont'))
```

```python
multiplication_sum = 0
```

```python
for i, mul in result_pairs:
    for j, command in combined_indices:
        if j > i:
            if command == 'dont':
                print('----------------')
                print('i: ', i)
                print('j: ', j)
                print('mul: ', mul)
                nums = re.findall( r'mul\((.*?),(.*?)\)', mul)[0]
                a = int(nums[0])
                b = int(nums[1])
                product = a * b
                multiplication_sum += product
            break
```
